```python
# create a scatter plot using the 'ra' and 'dec' columns, as color coding use mjd_first
plt.figure(figsize=(10, 6))
plt.scatter(df['sc_ra'], df['sc_dec'], c=df['mjd_first'], s=1, alpha=0.5)
plt.title('Scatter plot of RA vs DEC')
plt.xlabel('Right Ascension (RA)')
plt.ylabel('Declination (DEC)')
plt.colorbar(label='MJD First')
plt.grid(True)
plt.show()
```

```python
# create distribution plot for 'mjd_first'
plt.figure(figsize=(10, 6))
plt.hist(df['mjd_first'], bins=200, alpha=0.5)
plt.title('Distribution plot of MJD First')
plt.xlabel('MJD First')
plt.grid(True)
plt.show()
```
