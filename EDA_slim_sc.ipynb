```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('4xmm_slim_dr14_preprocessed.csv', index_col=0)
df
```

```python
df.columns
```

```python
feature_columns = [
    'sc_ra', 'sc_dec', 'sc_ep_1_flux', 'sc_ep_2_flux', 'sc_ep_3_flux',
       'sc_ep_4_flux', 'sc_ep_5_flux', 'sc_ep_8_flux', 'sc_ep_9_flux',
       'sc_ep_8_fmin', 'sc_ep_8_fmax', 'sc_hr1', 'sc_hr2', 'sc_hr3', 'sc_hr4',
       'sc_extent', 'sc_ext_ml', 'sc_chi2prob', 'sc_fvar', 'n_detections'
]
```

```python
# correlation analysis
correlation_matrix = df[feature_columns].corr()
plt.figure(figsize=(12, 10))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(feature_columns)), feature_columns, rotation=45, ha='right')
plt.yticks(range(len(feature_columns)), feature_columns)
plt.title('Correlation Matrix of Features')
plt.tight_layout()
plt.show()
```

```python
# create a scatter plot using the 'ra' and 'dec' columns, as color coding use mjd_first
plt.figure(figsize=(10, 6))
plt.scatter(df['sc_ra'], df['sc_dec'], c=df['mjd_first'], s=1, alpha=0.5)
plt.title('Scatter plot of RA vs DEC')
plt.xlabel('Right Ascension (RA)')
plt.ylabel('Declination (DEC)')
plt.colorbar(label='MJD First')
plt.grid(True)
plt.show()
```

```python
# create distribution plot for 'mjd_first'
plt.figure(figsize=(10, 6))
plt.hist(df['mjd_first'], bins=200, alpha=0.5)
plt.title('Distribution plot of MJD First')
plt.xlabel('MJD First')
plt.grid(True)
plt.show()
```

```python
# plot distribution for each feature column
for column in feature_columns:
    plt.figure(figsize=(10, 6))
    plt.hist(df[column], bins=200, alpha=0.5)
    plt.title(f'Distribution plot of {column}')
    plt.xlabel(column)
    plt.grid(True)
    plt.show()
```

```python
# count the rows with the same value in 'sc_ra' or 'sc_dec' columns approximating to n decimal places
for n in [0, 1, 2, 3, 4, 5, 6]:
    ra_counts_rounded = df['sc_ra'].round(n).value_counts()
    dec_counts_rounded = df['sc_dec'].round(n).value_counts()
    print(f"Average collision for (RA, DEC) - rounded to {n} decimal places: {float(ra_counts_rounded.mean()), float(dec_counts_rounded.mean())}")
```
